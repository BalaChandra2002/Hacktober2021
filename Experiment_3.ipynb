```python
import cv2
import numpy as np
import pandas as pd
```

```python
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/unnamed.jpg',0)
neg = 255 - img
```

```python
c = 255/np.log(1+np.max(img))
log = c*np.log(1+img+0.001)

gamma = np.power(img/255.0,2.5)*255
```

```python
cv2_imshow(neg)
cv2_imshow(log.astype(np.uint8))
cv2_imshow(gamma.astype(np.uint8))
```
